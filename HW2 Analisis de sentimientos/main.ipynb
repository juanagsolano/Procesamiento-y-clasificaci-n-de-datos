```python
import nltk
from nltk.corpus import stopwords
nltk.download("punkt")
nltk.download("stopwords")
nltk.download("wordnet")
```

### Preprocessing text aio

```python
import re
from nltk.corpus import wordnet
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer

wordnet_lemmatizer = WordNetLemmatizer()
stop = set(stopwords.words("english"))

pos_dict = {"J": wordnet.ADJ, "V": wordnet.VERB,
            "N": wordnet.NOUN, "R": wordnet.ADV}

def preprocesamiento_words(texto):
    """Preprocesamiento de palabras, retorna un listado de palabras con contexto.
    Retorna
    
    >>> Tupla (palabra,contexto)"""
    
    # Palabras a minusculas
    texto = texto.lower()
    # Eliminación de páginas web (palabras que empiezan por "http")
    texto = re.sub('http\S+', ' ', texto)
    # Removing Punctuations u otros caracteres
    texto = re.sub(r'[¿|?|!|\'|\"|#|:|“|”|_*]', r' ', texto)
    texto = re.sub(r'[.|,|)|(|\|/|\**|\-*]', r' ', texto)
    texto = re.sub(r'[\$\%\&\+\.\;\<\=\>\@\[\]\^_\`\{\|\}\~]', r' ', texto)
    # Eliminación de números
    texto = re.sub("\d+", ' ', texto)
    # Normaliza los espacios
    texto = re.sub(r' +', r' ', texto)

    # Separa oracion en palabras, remueve stopwords
    words = [word for word in texto.split() if word not in stop]
    
    # Añade contexto a las palabras
    tags = pos_tag(words)
    newlist = []
    for word, tag in tags:
        newlist.append(tuple([word,pos_dict.get(tag[0])]))    
    return newlist

def lemmatize(texto):
    """Retorna una lista de palabras lemmatizadas con contexto"""
    # Procesa y añade contexto a las palabras
    pos_data = preprocesamiento_words(texto)
    
    # Lematiza las palabras por contexto
    lemma_rew = []
    for word, pos in pos_data:
        if not pos:
            lemma_rew.append(word)
        else:
            lemma_rew.append(wordnet_lemmatizer.lemmatize(word, pos=pos))
    return lemma_rew

# Test function
lemmatize("https://sitioweb.com Nombre: Juan, Correo: juan.agsolano@gmail.com")
```

```python
new_df["Review_Token"] = new_df["Review"].apply(lemmatize)
new_df
```

### Lexicon for sentiment analysis

```python
lexicon = pd.read_table(
    'https://raw.githubusercontent.com/fnielsen/afinn/master/afinn/data/AFINN-en-165.txt',
    names=['termino', 'sentimiento'])
lexicon.rename(columns={"sentimiento":"lexicon sentiment"}, inplace=True)
lexicon.head()
```

```python
df_sent = pd.merge(left=explode_df,
                   right=lexicon,
                   left_on="Token",
                   right_on="termino",
                   how="left")
df_sent.drop("termino", axis=1, inplace=True)
df_sent
```

### To binary and categorical functions

```python
def binary_column(value):
    if value > 0:
        value = 1
    elif value < 0:
        value = 0
    else:
        value = 2
    return value

def categorical_column(value):
    if value > 0:
        value = "Positive"
    elif value < 0:
        value = "Negative"
    else:
        value = "Neutral"
    return value
```

### df sentiment by review

```python
review_sentiment =  df_sent.groupby(["id","Review_Date","Hotel_Name","Sentiment","Reviewer_Nationality"]).sum()
review_sentiment.reset_index(inplace=True)

review_sentiment["lexicon sentiment"] = review_sentiment["lexicon sentiment"].apply(lambda x: int(x))
review_sentiment["lexicon analysis"] = review_sentiment["lexicon sentiment"].apply(categorical_column)
review_sentiment
```

### Sentiment Proportions

```python
import numpy as np

def positive(x):
    return np.round(sum(x > 0)/len(x)*100,2)

def negative(x):
    return np.round(sum(x < 0)/len(x)*100,2)
        
def neutral(x):
    return np.round(sum(x == 0)/len(x)*100,2)
```

```python
sentiment_proportion_hotel = review_sentiment.groupby("Hotel_Name").agg({"lexicon sentiment":["count",positive,negative,neutral]})
sentiment_proportion_hotel = sentiment_proportion_hotel["lexicon sentiment"]
# Top 10 positive hotels
sentiment_proportion_hotel.sort_values(by=["count","positive"], ascending=False).iloc[:10,:]
```

```python
sentiment_prop_nat = review_sentiment.groupby("Reviewer_Nationality").agg(
    {"lexicon sentiment": ["count",positive, negative, neutral]})
sentiment_prop_nat = sentiment_prop_nat["lexicon sentiment"]
# Top 10 positive hotels
sentiment_prop_nat.sort_values(
    by=["count", "positive"], ascending=False).iloc[:10, :]
```

### Testing model

```python
# categorical columns estimate for model
import numpy as np
y_estimate = review_sentiment["lexicon sentiment"].apply(binary_column)
print("Estimated labels: {Negative: 0, Positive: 1, Neutral: 2}")

# binary sentiment column
y_real = review_sentiment["Sentiment"].factorize()
print("Sentiment labels:",y_real[1])
y_real = y_real[0]

# reviews size
n = len(y_real)

print("\nTesting Model")
# Proportion labels
print("Proportion: Positive/Total")
print(np.sum(review_sentiment["Sentiment"] == "Positive")/n*100)

# Accuraccy
print("\nAccuraccy")
print(np.sum(np.abs(y_estimate == y_real))/n*100)
```

## Sentiment Analysis using TextBlob

```python
from textblob import TextBlob

def getSubjectivity(review):
    return TextBlob(review).sentiment.subjectivity

def getPolarity(review):
    return TextBlob(review).sentiment.polarity

def analysis(score):
    if score < 0:
        return "Negative"
    elif score == 0:
        return "Neutral"
    else:
        return "Positive"
```

```python
fin_data["TextBlob Polarity"] = fin_data["lemma"].apply(getPolarity)
fin_data["TextBlob Analysis"] = fin_data["TextBlob Polarity"].apply(analysis)
fin_data
```

```python
# categorical columns estimate for model
import numpy as np
y_estimate = fin_data["TextBlob Polarity"].apply(binary_column)
print("Estimated labels: {Negative: 0, Positive: 1, Neutral: 2}")

# binary sentiment column
y_real = fin_data["Sentiment"].factorize()
print("Sentiment labels:", y_real[1])
y_real = y_real[0]

# reviews size
n = len(y_real)

print("\nTesting Model")
# Proportion labels
print("Proportion: Positive/Total")
print(np.sum(review_sentiment["Sentiment"] == "Positive")/n*100)

# Accuraccy
print("\nAccuraccy")
print(np.sum(np.abs(y_estimate == y_real))/n*100)
```

# Sentiment analysis with VADER

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()

def vader_analysis(text):
    vs = analyzer.polarity_scores(text)
    return vs["compound"]

fin_data["Vader Sentiment"] = fin_data["lemma"].apply(vader_analysis)
```

```python
def vader_categorical(value):
    if value >= 0.5:
        return "Positive"
    elif value <= 0.5:
        return"Negative"
    else:
        return "Neutral"
    
fin_data["Vader Analysis"] = fin_data["Vader Sentiment"].apply(vader_categorical)
```

# Evaluating models

```python
def label_to_int(label):
    if label == "Positive":
        x = 1
    elif label == "Negative":
        x = 0
    else:
        x = 1
    return x
```

```python
from sklearn.metrics import classification_report, confusion_matrix, \
f1_score, accuracy_score,recall_score,precision_score
import seaborn as sns

y_true = fin_data["Sentiment"].apply(label_to_int)
f1 = []
Ac = []
Rc = []
Pr = []

# Lexicon model
y_pred = fin_data["lexicon analysis"].apply(label_to_int)

print("Performance Model Report\n")
print(classification_report(y_true,y_pred))

f1.append(f1_score(y_true, y_pred)*100)
Ac.append(accuracy_score(y_true,y_pred)*100)
Rc.append(recall_score(y_true,y_pred)*100)
Pr.append(precision_score(y_true,y_pred)*100)

print("f1-score: {:4.2f}%".format(f1[0]))
print("Accuracy: {:4.2f}%".format(Ac[0]))
print("Recall: {:4.2f}%".format(Rc[0]))
print("Precision: {:4.2f}%".format(Pr[0]))

# Create and plot confusion matrix
cm = confusion_matrix(y_true,y_pred)/len(y_true)
plt.figure(figsize=(8,6))
ax = sns.heatmap(cm, annot = True, cmap="Blues", fmt="0.2%")
ax.set_title("Confusion Matrix for Lexicon Model")

ax.set_xlabel("Predicted Values")
ax.set_ylabel("True Values")

ax.xaxis.set_ticklabels(["False","True"])
ax.yaxis.set_ticklabels(["False","True"])
plt.show()
```
